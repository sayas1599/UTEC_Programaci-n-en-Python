```python
# Creación de la función de limpieza de tablas

def limpieza_tabla(tabla):
  """ Esta función realiza la limpieza de tablas, eliminando filas y columnas vacías, y ajustando el índice de tabla """
  # Función iloc para hacer un filtro de los datos por posición
  tabla = tabla.iloc[3:]
  cabeceras = tabla.iloc[0]
  tabla.columns = cabeceras
  tabla = tabla.iloc[1:]
  tabla.reset_index(drop=True, inplace=True)
  # Eliminando las columnas totalmente vacías
  tabla = tabla.dropna(axis = 1,how= "all")
  return tabla
```

```python
# Aplicación de la función limpieza_tabla para las 04 hojas de cálculo

paradas_molinocarbon = limpieza_tabla(paradas_molinocarbon)
paradas_crudo = limpieza_tabla(paradas_crudo)
paradas_horno = limpieza_tabla(paradas_horno)
paradas_molino = limpieza_tabla(paradas_molino)
```

```python
# 1. Tiempo total de paradas en minutos por Sección, Año y Mes.
data_final.pivot_table(values='Tiempo Min', index=['AÑO', 'MES'],columns='Sección',aggfunc='sum')
```

```python
# 2. Tiempo total de paradas en minutos por Causa de parada y año.

data_final.pivot_table(values='Tiempo Min', index=['AÑO','Causa'],columns='Sección',aggfunc='sum')
```

```python
# 3. Tiempo total de paradas por Tipo de paradas (Programada o no programada) por año y sección
data_final.pivot_table(values='Tiempo Min', index=['AÑO','Tipo '],columns='Sección',aggfunc='sum')
```

```python
# 4. Tiempo máximo de paradas por Tipo de paradas (Programada o no programada) por año y sección
data_final.pivot_table(values='Tiempo Min', index=['AÑO', 'Tipo '],columns='Sección',aggfunc='max')
```

```python
# 5. Mediana del tiempo de paradas en minutos por Causa de parada y año.
Sección_causa_tiempo = data_final.groupby(['Sección', 'Causa'])['Tiempo Min'].median().reset_index ()
Sección_causa_tiempo = Sección_causa_tiempo.sort_values(by=['Sección', 'Tiempo Min'], ascending=[True, False])

print("Impacto de las Causas en el Tiempo de Parada por Equipo")
print(Sección_causa_tiempo)
```

```python
# 6. Frecuencia de paradas por Sección y Clase de paradas.

frecuencia_paradas = data_final.groupby(['Sección', 'Causa']).size().reset_index(name='Frecuencia')

print("Frecuencia de Paradas por Sección y Clase")
print(frecuencia_paradas)
```

```python
# 7. Tiempo total de paradas por Sub Clase y por año.

data_final = data_final.dropna(subset=['Sub Clase '])
data_final.pivot_table(values='Tiempo Min', index=['AÑO','Sub Clase '],columns='Sección',aggfunc='sum')
```

```python
# Definición de la función de obtención de tiempo de paradas por Sub Clase, sección, mes y año.
def obtener_paradas(data, seccion, año, mes):

    filtro = (data['Sección'] == seccion) & (data['AÑO'] == año) & (data['MES'] == mes)
    datos_filtrados = data[filtro]

    if datos_filtrados.empty:
        return f"No hay datos disponibles para la sección {seccion} en {año}-{mes}."

    resultados = datos_filtrados[['Tiempo Min', 'Sub Clase ']]

    return resultados

print (f"Los resultado de la sección {seccion} en {año}-{mes} son: .")
obtener_paradas(data_final, seccion, año, mes)
```
