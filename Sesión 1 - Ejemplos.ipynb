## 1. Media / Promedio

Es el valor promedio de un conjunto de datos, calculado sumando todos los valores y dividiendo por el número total de valores.

```python
import numpy as np

# Datos de ejemplo
datos = [5, 10, 15, 20, 25]

# Calcular la media
media = np.mean(datos)

print("Media:", media)
```

## 2. Mediana

Es el valor que se encuentra en el centro de un conjunto de datos cuando están ordenados de menor a mayor, dividiendo el conjunto en dos partes iguales.



```python
import numpy as np

# Datos de ejemplo
datos = [5, 10, 15, 20, 25]

# Calcular la mediana
mediana = np.median(datos)

print("Mediana:", mediana)
```

## 3. Máximo y mínimo


```python
import numpy as np

# Datos de ejemplo
datos = [5, 10, 15, 20, 25]

# Calcular el máximo y el mínimo
maximo = np.max(datos)
minimo = np.min(datos)

print("Máximo:", maximo)
print("Mínimo:", minimo)
```

## 4. Moda
 Es el valor que aparece con mayor frecuencia en un conjunto de datos.



```python
import pandas as pd

# DataFrame con datos categóricos y numéricos
datos = {
    'Persona': ['Ana', 'Luis', 'Juan', 'Maria', 'Pedro', 'Laura', 'Elena', 'Carlos', 'Ana', 'Luis'],
    'Color Preferido': ['Azul', 'Rojo', 'Verde', 'Azul', 'Azul', 'Rojo', 'Verde', 'Rojo', 'Azul', 'Azul'],
    'Edad': [23, 34, 23, 45, 23, 34, 34, 45, 23, 34]
}

df = pd.DataFrame(datos)

# Mostrar las primeras filas del DataFrame
print("Datos de entrada:")
print(df)

# Calcular la moda para la columna 'Color Preferido'
moda_color_preferido = df['Color Preferido'].mode()[0]

# Mostrar la moda
print("\nEl color más preferido es:", moda_color_preferido)


# Calcular la moda para la columna 'Edad'
moda_edad = df['Edad'].mode()[0]

# Mostrar la moda
print("\nLa edad más común es:", moda_edad)
```

## **5. Desviación estándar**
Es una medida de dispersión que indica cuánto varían los valores de un conjunto de datos con respecto a la media.

```python
import numpy as np

# Datos de ejemplo
datos = [5, 10, 15, 20, 25]

# Calcular la desviación estándar
desviacion_estandar = np.std(datos)

print("Desviación estándar:", desviacion_estandar)
```

## **6. Correlación**
Es una medida estadística que describe la relación entre dos variables. Indica la fuerza y la dirección de la relación lineal entre ellas.

image.png

```python
import pandas as pd

# Datos de ejemplo
datos = {
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 6, 8, 10]
}

# Convertir a DataFrame
df = pd.DataFrame(datos)

# Calcular la correlación
correlacion = df['x'].corr(df['y'])

print("Correlación:", correlacion)
```

```python
import matplotlib.pyplot as plt

# Graficar los datos
plt.scatter(df['x'], df['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de correlación de X vs Y')
plt.show()
```

## 7. **Quantil superior e inferior**

Son valores que dividen un conjunto de datos ordenados en partes iguales, donde el quantil superior representa el valor por encima del cual se encuentran cierto porcentaje de los datos y el quantil inferior representa el valor por debajo del cual se encuentran cierto porcentaje de los datos.

image.png


image.png

```python
import numpy as np

# Datos de ejemplo
datos = [5, 10, 15, 20, 25]

# Calcular quantiles
quantil_superior = np.percentile(datos, 75)
quantil_inferior = np.percentile(datos, 25)

print("Quantil superior:", quantil_superior)
print("Quantil inferior:", quantil_inferior)
```

```python
import matplotlib.pyplot as plt

# Graficar los datos
plt.boxplot(datos)
plt.title('Diagrama de caja y bigotes de los datos')
plt.ylabel('Valor')
plt.show()
```

Ejemplo con datos de calificaciones

```python
import pandas as pd
import numpy as np

# Datos de calificaciones
calificaciones = {
    'Estudiante': ['Ana', 'Luis', 'Juan', 'Maria', 'Pedro', 'Laura', 'Elena', 'Carlos', 'Julia', 'Pablo'],
    'Calificación': [85, 90, 78, 92, 88, 76, 95, 89, 84, 91]
}

df = pd.DataFrame(calificaciones)

# Calcular el cuantil inferior (Q1) y el cuantil superior (Q3)
Q1 = df['Calificación'].quantile(0.25)
Q3 = df['Calificación'].quantile(0.75)

# Mostrar los cuantiles
print("Calificaciones:")
print(df)
print(f"\nEl cuantil inferior (Q1) es: {Q1}")
print(f"El cuantil superior (Q3) es: {Q3}")
```

```python
import matplotlib.pyplot as plt

# Graficar los datos
plt.boxplot(df['Calificación'])
plt.title('Diagrama de caja y bigotes de Calificaciones')
plt.ylabel('Valor')
plt.show()
```
