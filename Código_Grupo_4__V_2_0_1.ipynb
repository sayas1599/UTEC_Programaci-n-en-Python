```python
def limpieza_tabla(tabla):
  tabla = tabla.iloc[3:]
  cabeceras = tabla.iloc[0]
  tabla.columns = cabeceras
  tabla = tabla.iloc[1:]
  tabla.reset_index(drop=True, inplace=True)
  tabla = tabla.dropna(axis = 1,how= "all")
  return tabla

```

```python
def obtener_paradas(data, seccion, año, mes):
    # Filtrar los datos por sección, año y mes
    filtro = (data['Sección'] == seccion) & (data['AÑO'] == año) & (data['MES'] == mes)
    datos_filtrados = data[filtro]

    # Verificar si hay datos para los filtros aplicados
    if datos_filtrados.empty:
        return f"No hay datos disponibles para la sección {seccion} en {año}-{mes}."

    # Seleccionar las columnas relevantes
    resultados = datos_filtrados[['Tiempo Min', 'Causa']]

    return resultados

obtener_paradas(data_final, 'Horno Rotativo 1', 2023, 1)
```

```python
def obtener_detalle_paradas(data, seccion, año, mes):

    filtro_seccion = (data['Sección'] == seccion) & (data['AÑO'] == año) & (data['MES'] == mes)
    datos_filtrados = data.loc[filtro_seccion]

    if datos_filtrados.empty:
        return f"No hay datos disponibles para la sección {seccion} en {año}-{mes}."

    detalles_paradas = datos_filtrados[['Tiempo Min', 'Causa']]

    return detalles_paradas
```
