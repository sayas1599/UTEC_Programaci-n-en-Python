# Tarea : Leer los 3 archivos restantes

```python
def limpieza_tabla(tabla):
  tabla = tabla.iloc[3:]
  cabeceras = tabla.iloc[0]
  tabla.columns = cabeceras
  tabla = tabla.iloc[1:]
  tabla.reset_index(drop=True, inplace=True)
  tabla = tabla.dropna(axis = 1,how= "all")
  return tabla

```

```python
paradas_embolsado1 = pd.read_excel("/content/drive/MyDrive/2024 - Grupo 4 -Trabajo Grupal - Python (UTEC)/Segundo Entregable/Datos/Paradas Embolsado1 2023-2024.xlsx")
paradas_embolsado2 = pd.read_excel("/content/drive/MyDrive/2024 - Grupo 4 -Trabajo Grupal - Python (UTEC)/Segundo Entregable/Datos/Paradas Embolsado2 2023-2024.xlsx")
paradas_horno = pd.read_excel("/content/drive/MyDrive/2024 - Grupo 4 -Trabajo Grupal - Python (UTEC)/Segundo Entregable/Datos/Paradas Horno 2023-2024.xlsx")
paradas_molino = pd.read_excel("/content/drive/MyDrive/2024 - Grupo 4 -Trabajo Grupal - Python (UTEC)/Segundo Entregable/Datos/Paradas Molino de Cemento 2023-2024.xlsx")
```
