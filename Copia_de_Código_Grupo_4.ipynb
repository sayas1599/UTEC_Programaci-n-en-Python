# Tarea : Leer los 3 archivos restantes

```python
def limpieza_tabla(tabla):
  tabla = tabla.iloc[3:]
  cabeceras = tabla.iloc[0]
  tabla.columns = cabeceras
  tabla = tabla.iloc[1:]
  tabla.reset_index(drop=True, inplace=True)
  tabla = tabla.dropna(axis = 1,how= "all")
  return tabla

```
